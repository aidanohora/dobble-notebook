-The below cell defines a function to check the validity of the 'deck' (the dictionary which holds all potential cards).

-It compares every card in dictionary with every other card in it at least once to determine that they all have no more and no less than one image in common.

-It can be called with an optional key word arguement (verbose = True) in order to produce output every time it compares two cards. 

```python
def check_validity(deck, **kwargs):
    invalid = False #will be switched to true if the deck is invalid
    intersectionlist = [] #will hold all the numbers in common between two cards 
    if 'verbose' in kwargs and kwargs['verbose']==True:
        print("Comparing cards images on cards...")
    for i in range(1, 58): #for each card in the deck of 57
        for j in range(i, 58): #starts at i to avoid repeating comparisons of the same cards
            if newdict[i] != newdict[j]: #so that cards that are the same aren't compared
                intersectionlist.append(newdict[i].intersection(newdict[j])) #appending shared numbers to intersectionlist
                if len(intersectionlist) != 1: 
                    if 'verbose' in kwargs and kwargs['verbose']==True:
                        if len(intersectionlist) > 1:
                            print("Error: Cards match on more than one image!")
                            invalid = True
                        else:
                            print("Error: Cards don't match on any images!")
                            invalid = True
                elif 'verbose' in kwargs and kwargs['verbose']==True:
                    print("Last two cards compared match on only one image.")
            intersectionlist = [] #emptying intersectionlist before the next comparison
        if 'verbose' in kwargs and kwargs['verbose']==True:
            if invalid==False:
                print()
                print("Deck is valid - any two cards in this deck match on one and only one image.")
                print()
                return
            if invalid==True:
                print()
                print("This deck is invalid.")
                print()
                return
```

-The cells below include code defining the DobbleCard class, the DobbleDeck class (and it's add_card, remove_card and play_card methods) and a play_dobble method which will simulate a game of dobble.

-I have attempted to include all the neccesary error handling that the player will always be prompted for input again when invalid input is entered, with an explanation of why the input was invalid.

-There is a quit feature during the playing of the cards, so if you decide to do 27 rounds but want to end the game early, instead of typing 'a', 'b' or 'd' for who won, type 'q' or 'Q'. 

```python
from random import choice 

class DobbleCard:
    keys = list(range(1, 58))
    
    def __init__(self):
        self.images = newdict[choice(DobbleCard.keys)] #randomly taking a set from newdict for card

class DobbleDeck:
    indeck = [] #this holds the DobbleCard objects
    cards = 0 #keeps track of the number of cards in the deck, basically len(DobbleDeck object.indeck)

    def add_card(self, toadd): 
        '''Creates a card object and adds it to a DobbleDeck objects indeck list
            second arguement takes an integer value to determine how many cards are created/added'''
        while toadd > 0:
            card = DobbleCard()
            if card not in self.indeck: #will try another card if that particular card has been added
                self.indeck.append(card)
                toadd -= 1
                self.cards += 1

    def remove_card(self, toremove):
        '''removes a card object from a DobbleDeck objects indeck list
            second arguement decides the index in the list for the card to be removed'''
        i = toremove
        self.indeck.pop(i)
        self.cards -= 1

    def play_card(self):
        '''displays the emojis of the two cards at the front of a DobbleDeck objects indeck list'''
        if len(self.indeck) >= 2: #one card cannot be played on it's own
            cardA = self.indeck[0] #taking/removing the 'top' card from the deck
            DobbleDeck.remove_card(self, 0) 
            cardB = self.indeck[0] #taking/removing the 'top' card from the deck a second time
            DobbleDeck.remove_card(self, 0)
            displayA = list(cardA.images) #creating two variables for displaying the                 
            for i in range(0, 8):         #emojis corresponding to the keys of the two cards
                number = displayA[i]
                displayA[i] = imageDict[number]
            displayB = list(cardB.images)
            for i in range(0, 8):
                number = displayB[i]
                displayB[i] = imageDict[number]
            print(displayA[0], displayA[1], displayA[2],"     ", displayB[0], displayB[1], displayB[2])
            print(displayA[3], displayA[4], displayA[5],"     ", displayB[3], displayB[4], displayB[5])
            print(displayA[6], displayA[7], "         ", displayB[6], displayB[7]) #displaying cards
        else:
            print("Error: you need at least two cards left to play a round.")

def play_dobble():
    '''A function to create a DobbleDeck object and use it to play a game of dobble,
        relying on the add_card, remove_card and playcard functions'''
    Apoints = 0
    Bpoints = 0
    validcardno = False 
    deck = DobbleDeck() #creating the deck that will be used for this game
    while validcardno == False : 
        cardstouse = input("How many cards (<56)?") #players will be asked until valid input is given
        try:
            cardstouse = int(cardstouse)
        except ValueError:
            print("Please only enter an integer value for cards.")
            continue
        else:    
            if cardstouse >= 56:
                print("Please choose less than 56 cards.")
            elif cardstouse < 2:
                print("You cannot play with less than 2 cards.")
            else:
                validcardno = True #loop can be exited, valid input has been received
    deck.add_card(cardstouse)
    print("If you want to record a draw type 'd' or 'D', or if you want to end the game early type 'q' or 'Q'")
    print()
    while deck.cards > 1: #cards will be drawn/played until there are no longer two cards in the deck
        DobbleDeck.play_card(deck)
        validinput = False 
        while validinput != True: #loop is exited once valid input is received
            winner = str(input("Who wins (A or B)?"))
            print()
            winner = winner.upper()
            if winner == "A": 
                Apoints += 1
                validinput = True
            elif winner == "B":
                Bpoints += 1
                validinput = True
            elif winner == "D":
                validinput = True
            elif winner == "Q": #emergency quit option
                validinput = True
                deck.cards = 0 #ends the game early
            else:
                print("Please only enter A or B.")
                continue
    if deck.cards == 1:
        print()
        print("Only one card left in the deck - end of game.")
    elif deck.cards == 0:
        print()
        print("End of game.")
    print()
    print("Score")
    print("A:", Apoints)
    print("B:", Bpoints)
    #game is complete

                
        
        
        
            
            
 
        
            
            
```
